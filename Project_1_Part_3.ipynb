```python
#Import libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
#data loaded
foodsales = "/content/sales_predictions.csv"
df = pd.read_csv(foodsales)
df.head()
```

# Histogram to view the distributions of various features in your dataset.

```python
plt.style.use('seaborn')
fig, ax = plt.subplots(figsize=(10, 6))
ax.hist(df["Item_Type"], bins="auto", edgecolor="black");
plt.xlabel("Item_type", fontsize=20)
plt.ylabel("Count", fontsize=20)
plt.xticks(fontsize=16,rotation = 90)
plt.yticks(fontsize = 20)
```

```python
df["Item_Fat_Content"] = df["Item_Fat_Content"].replace("LF", "Low Fat")
df["Item_Fat_Content"] = df["Item_Fat_Content"].replace("reg", "Regular")
df["Item_Fat_Content"] = df["Item_Fat_Content"].replace("low fat", "Low Fat")
```

```python
plt.style.use('seaborn')
fig, ax = plt.subplots(figsize=(10, 6))
ax.hist(df["Item_Fat_Content"], bins="auto", edgecolor="black");
plt.xlabel("Item_Fat_Content", fontsize=20)
plt.ylabel("Count", fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize = 20)
```

```python
average_weight = df["Item_Weight"].mean()
df["Item_Weight"].fillna(average_weight, inplace=True)
```

```python
plt.style.use('seaborn')
fig, ax = plt.subplots(figsize=(10, 6))
ax.hist(df["Item_Weight"], bins="auto", edgecolor="black");
ax.hist(df["Item_Type"], bins="auto", edgecolor="black");
plt.xlabel("Item_Type & Item_Weight", fontsize=20)
plt.ylabel("Count", fontsize=20)
plt.xticks(fontsize=16,rotation = 90)
plt.yticks(fontsize = 16)
```

# Boxplot to view statistical summaries of various features in your dataset.

```python
plt.style.use('seaborn')
df.boxplot(column = "Item_Visibility", by = "Item_Type")
plt.xlabel("Item_Type", fontsize=20)
plt.ylabel("Item_Visibility", fontsize=20)
plt.xticks(fontsize=16,rotation = 90)
plt.yticks(fontsize = 16)
plt.title("The Item Visibility of Each Item Type", fontsize = 20 )
```

```python
plt.style.use('seaborn')
df.boxplot(column = "Item_Outlet_Sales", by = "Outlet_Location_Type")
plt.xlabel("Outlet_Location_Type", fontsize=20)
plt.ylabel("Item_Outlet_Sales", fontsize=20)
plt.xticks(fontsize=16)
plt.yticks(fontsize = 16)
plt.title("The product sales of the type of area store located", fontsize = 20 )
```

```python
plt.style.use('seaborn')
df.boxplot(column = "Item_MRP", by = "Item_Type")
plt.xlabel("Item_type", fontsize=20)
plt.ylabel("Item_MRP", fontsize=20)
plt.xticks(fontsize=16,rotation = 90)
plt.yticks(fontsize = 16)
plt.title("The MRP of Each Type of Item", fontsize = 20 )
```

# Heatmap of the correlation between features

```python
corr = df.corr()
plt.figure(figsize=(10,10))
sns.heatmap(corr, cmap="Oranges", annot=True)
plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20, rotation = 360)
```
